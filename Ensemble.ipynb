```python
import pandas as pd
```

```python
monologg_kobigbird_v0 = pd.read_csv('monologg_kobigbird_v0.csv')
monologg_kobigbird_v1 = pd.read_csv('bigbird_v3.csv')
monologg_koelectra_v0 = pd.read_csv('koelectra_v0(monolog).csv')
monologg_koelectra_v1 = pd.read_csv('koelectra_base_aihub_v2.csv')
kykim_electra_v0 = pd.read_csv('kykim_electra_v0.csv')
kykim_electra_v1 = pd.read_csv('koelectra_base_aihub_kykim.csv')
tunib_electra_v0 = pd.read_csv('tunib_electra_v0.csv')
tunib_electra_v1 = pd.read_csv('tunib_electra_v3.csv')
xml_roberta_v0 =  pd.read_csv('xml_roberta_v0.csv')
```

```python
monologg_kobigbird_v0.rename(columns={'Predicted':'Predicted1'}, inplace = True)
monologg_kobigbird_v1.rename(columns={'Predicted':'Predicted2'}, inplace = True)

monologg_koelectra_v0.rename(columns={'Predicted':'Predicted3'}, inplace = True)
monologg_koelectra_v1.rename(columns={'Predicted':'Predicted4'}, inplace = True)

kykim_electra_v0.rename(columns={'Predicted':'Predicted5'}, inplace = True)
kykim_electra_v1.rename(columns={'Predicted':'Predicted6'}, inplace = True)

tunib_electra_v0.rename(columns={'Predicted':'Predicted7'}, inplace = True)
tunib_electra_v1.rename(columns={'Predicted':'Predicted8'}, inplace = True)

xml_roberta_v0.rename(columns={'Predicted':'Predicted9'}, inplace = True)
```

```python
data = pd.merge(monologg_kobigbird_v0,monologg_kobigbird_v1, on='id')
data = pd.merge(data,monologg_koelectra_v0, on='id')
data = pd.merge(data,monologg_koelectra_v1, on='id')
data = pd.merge(data,kykim_electra_v0, on='id')
data = pd.merge(data,kykim_electra_v1, on='id')
data = pd.merge(data,tunib_electra_v0, on='id')
data = pd.merge(data,tunib_electra_v1, on='id')
predict_data = pd.merge(data,xml_roberta_v0, on='id')
```

```python
predict_data
```

```python
from collections import Counter

def find_max(word):
    counter = Counter(word)
    max_count = -1
    for letter in counter:
        if counter[letter] > max_count:
            max_count = counter[letter]
            max_letter = letter
    return max_letter, max_count

for i in range(len(predict_data)):
    candidate = []
    candidate.append(predict_data.loc[i, 'Predicted1'])
    candidate.append(predict_data.loc[i, 'Predicted2'])
    candidate.append(predict_data.loc[i, 'Predicted3'])
    candidate.append(predict_data.loc[i, 'Predicted4'])
    candidate.append(predict_data.loc[i, 'Predicted5'])
    candidate.append(predict_data.loc[i, 'Predicted6'])
    candidate.append(predict_data.loc[i, 'Predicted7'])
    candidate.append(predict_data.loc[i, 'Predicted8'])
    candidate.append(predict_data.loc[i, 'Predicted9'])
    ans, confidence = find_max(candidate)
    predict_data.loc[i, 'result'] = ans
    predict_data.loc[i, 'confidence'] = confidence
```

```python
predict_data.drop(['Predicted1', 'Predicted2','Predicted3','Predicted4','Predicted5','Predicted6','Predicted7','Predicted8','Predicted9'], axis=1, inplace=True)
```

```python
predict_data.rename({'result':'Predicted'}, axis='columns', inplace=True)
```

# 후처리

```python
ans_list = predict_data['Predicted'].values.tolist()
our_ans_conc = []
for ans in ans_list: #길이가 긴 값은 줄여버리기
    if type(ans) == float:
        our_ans_conc.append([''])
    elif len(ans) > 24:
        our_ans_conc.append(ans[0:5])
    else:
        our_ans_conc.append(ans)
our_ans_conc
```

```python
predict_data['Predicted'] = our_ans_conc
predict_data
```

```python
from numpy import nan

for i in range(len(predict_data)):
    if predict_data.loc[i,'confidence'] < 2:
        predict_data.loc[i,'ans'] = nan
    else:
        predict_data.loc[i,'ans'] = predict_data.loc[i,'Predicted']

predict_data
```

```python
predict_data.drop(['Predicted','confidence'], axis=1, inplace=True)
predict_data.rename({'ans':'Predicted'}, axis='columns', inplace=True)
predict_data
```

```python
predict_data.to_csv('ensemble.csv', index=False)
```
